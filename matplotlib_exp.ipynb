```
plt.figure(2,dpi=300)
plt.plot(x,y_1)
plt.plot(x,y,color='red',linewidth=1.0,linestyle='--')
plt.xlim((-1,2))
plt.ylim((-2,3))
plt.xlabel('x label')
plt.ylabel('y label')
new_ticks = np.linspace(-1,2,5)
print(new_ticks)
plt.xticks(new_ticks)   #增加的代码使得横坐标的坐标点变成五个点
plt.yticks([-2,-1.8,-1,1.22,3],
           [r'$really\ bad$', r'$bad$', r'$normal$', r'$good$', r'$really\ good$'])
#将对应值所表示的属性显示
ax = plt.gca()     #gca: get current axis
ax.spines['right'].set_color('none') #去除坐标轴的右边框
ax.spines['top'].set_color('none')  #去除坐标轴的上边框
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
#设置哪个轴为x轴哪个为y轴，因为plt中x轴可能是上框线也可能是下框线，所以需要人工设置
ax.spines['bottom'].set_position(('data',-1))  #横坐标的起始点是y轴的-1
ax.spines['left'].set_position(('data',0))
plt.show()
```

```
plt.figure(dpi=300)
plt.xlim((-1,2))
plt.ylim((-2,3))
plt.xlabel('x label')
plt.ylabel('y label')

new_ticks = np.linspace(-1,2,5)
print(new_ticks)
plt.xticks(new_ticks)   
plt.yticks([-2,-1.8,-1,1.22,3],
           [r'$really\ bad$', r'$bad$', r'$normal$', r'$good$', r'$really\ good$'])
#将对应值所表示的属性显示

l1, = plt.plot(x,y_1,label='up')
l2, = plt.plot(x,y,color='red',linewidth=1.0,linestyle='--',label='down')
plt.legend(handles=[l1,l2,],labels=['A','B'],loc='lower right')
plt.show()
```

有关为什么l1, = ... 以及handles=[l1,l2,]中这神奇的逗号，可以参考https://stackoverflow.com/questions/11983024/matplotlib-legends-not-working中的回答，与plt返回值的类型有关。

```
n = 1024
X = np.random.normal(0,1,n)  #均值为0方差为1的随机数
Y = np.random.normal(0,1,n)
T = np.arctan2(Y,X)      #用公式设置颜色，只是为了好看

plt.scatter(X,Y,s=75,c=T,alpha=0.5)

plt.xlim((-1.5,1.5))
plt.ylim((-1.5,1.5))

plt.xticks(())
plt.yticks(())

plt.show()
```

```
n_1 = 12
X_1 = np.arange(n_1)
Y_1 = (1-X_1/float(n_1))*np.random.uniform(0.5,1.0,n_1)
Y_2 = (1-X_1/float(n_1))*np.random.uniform(0.5,1.0,n_1)

plt.bar(X_1,+Y_1,facecolor='#9999ff',edgecolor='white')
plt.bar(X_1,-Y_2,facecolor='#ff9999',edgecolor='white')

#显示柱状图的值
for x,y in zip(X_1,Y_1):   #zip使得能向x,y传入X_1，Y_1两个值
  #设置文本对齐方式
  plt.text(x,y+0.05,'%.2f' % y,ha='center',va='bottom')

for x,y in zip(X_1,Y_2):   #zip使得能向x,y传入X_1，Y_1两个值
  plt.text(x,-y-0.05,'-%.2f'%y,ha='center',va='top')



plt.xlim((-1.25,n_1))
plt.ylim((-1.5,1.5))

plt.xticks(())
plt.yticks(())
```

```
def f(x,y):
  #计算海拔高度
  return (1 - x / 2 + x ** 5 + y ** 3)*np.exp(-x**2-y**2)

n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
X_2, Y_2 = np.meshgrid(x,y)

plt.contourf(X_2, Y_2, f(X_2,Y_2), 8, aplha=0.75, cmap=plt.cm.hot) #alpha为透明度

#绘制等高线
C = plt.contour(X_2, Y_2, f(X_2, Y_2), 8, color='black',linewidth=.5)
#8表示等高线一共有（8+2）种

#添加标签
plt.clabel(C,inline=True,fontsize=10)


plt.xticks(())
plt.yticks(())
plt.show()
```

```
from mpl_toolkits.mplot3d import Axes3D  #导入显示3D坐标的模块
fig = plt.figure(dpi=100)
ax = Axes3D(fig)
#设置横纵坐标值
X_3 = np.arange(-4,4,0.25)
Y_3 = np.arange(-4,4,0.25)
X_3, Y_3 = np.meshgrid(X_3, Y_3)
R = np.sqrt(X_3**2+Y_3**2)
#设置Z坐标值
Z_3 = np.sin(R)

ax.plot_surface(X_3, Y_3, Z_3, rstride=1, cstride=1, cmap=plt.get_cmap('rainbow'))
#rstride与xstride决定了经过多长绘制一种颜色，通过更改其值可明显看出差别
ax.contourf(X_3, Y_3, Z_3, zdir='z',offset=-2,cmap='rainbow')
ax.set_zlim(-2,2)
#offset=-2指将投影的图像放置于di'yu
plt.show()
```

```
fig = plt.figure(dpi=100)
x = [1,2,3,4,5,6,7]
y = [1,3,4,2,5,8,6]

left, bottom, width, height = 0.1, 0.1, 0.8, 0.8  #设定白色画的大小，即从figure左边10%处，底部10%处。。。。。为画布
ax1 = fig.add_axes([left, bottom, width, height])
ax1.plot(x,y,'r')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('title')
#method 1
left, bottom, width, height = 0.2, 0.6, 0.25, 0.25  #设定白色画的大小，即从figure左边10%处，底部10%处。。。。。为画布
ax2 = fig.add_axes([left, bottom, width, height])
ax2.plot(y,x,'b')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('insert_1')
#method 2
plt.axes([0.6,0.2,0.25,0.25])
plt.plot(y[::-1],x,'g')   #y[::-1]表示从前往后，间隔为-1，相当于倒序y
plt.xlabel('x')
plt.ylabel('y')
plt.title('insert_2')

plt.show()
```

动画形式展示数据，本节为正弦波动画

```
from IPython.display import HTML
from matplotlib import animation,rc
fig, ax = plt.subplots()

x = np.arange(0, 2*np.pi,0.01)
line, = ax.plot(x,np.sin(x))


def animate(i):  #当前帧数为i，调用时会自动传入
  line.set_ydata(np.sin(x+i/100))
  return line,   #因为ax.plot返回的是个列表，加逗号就表明接收的是第一个值

def init():
  line.set_ydata(np.sin(x))
  return line,   #因为ax.plot返回的是个列表，加逗号就表明接收的是第一个值



ani = animation.FuncAnimation(fig=fig, func=animate, frames=100, init_func=init, interval=20, blit=False)
#interval 更新频率，即多少毫秒更新一次
#blit=True只更新有变化的点
rc('animation', html='jshtml')
ani
```
